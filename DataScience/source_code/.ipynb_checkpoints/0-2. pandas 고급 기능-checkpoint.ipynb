### 1. 두 DataFrame 병합하기

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0,1,2,3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4,5,6,7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

* 열 방향으로 단순히 합치는 것은 concat 명령

```python
result = pd.concat([df1, df2, df3])
result
```

* concat 명령에 keys 옵션으로 구분할 수 있다.
* key 지정된 구분은 다중 index가 되어서 level을 형성

```python
result = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
result
```

```python
result.index
```

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2,3,6,7])

result = pd.concat([df1, df4], axis=1)
```

* concat 명령은 index를 기준으로 데이터를 합치기 때문에 값을 가질 수 없는 곳에는 NaN이 저장된다

```python
result
```

* 공통된 index로 합치고 공통되지 않은 index의 데이터는 버리도록 하는 옵션이 join='inner'

```python
result = pd.concat([df1, df4], axis=1, join='inner')

result
```

* join_axes=[df1.index] 옵션으로 df1의 인덱스에 맞추로독 할 수도 있다

```python
result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])

result
```

concat 명령을 사용하는데 열 방향으로 합치면서 ignore_index=True 라고 옵션을 걸면

두 데이터의 index를 무시하고 합친 후 다시 index를 부여한다

이때는 열을 기준으로 합치게 된다

* FutureWarning: Sorting because non-concatenation axis is not aligned. A future version of pandas will change to not sort by default.   가 발생해서 나는 아래 옵션을 추가했다
* sort=False

```python
result = pd.concat([df1, df4], ignore_index=True, sort=False)

result
```

```python
left = pd.DataFrame({'key': ['K0', 'K4', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

left와 right 두 데이터에 공통으로 있는 컬럼인 key를 기준으로

merge 명령에서 merge 기준을 설정하는 on 옵션으로 합치면 공통된 key에 대해서만 합치게 된다

```python
pd.merge(left, right, on='key')
```

* 합치는 두 데이터를 하나씩 기준으로 합칠 수도 있다. 이때 how 옵션으로 한쪽 데이터를 설정하면 된다.

```python
pd.merge(left, right, how='left', on='key')
```

```python
pd.merge(left, right, how='right', on='key')
```

* merge 한 데이터 결과를 모두 가지는 outer 옵션이 있다.
* 이 옵션을 사용하면 합집합처럼 merge가 되며 공통된 요소가 아닌 곳은 NaN 처리가 된다

```python
pd.merge(left, right, how='outer', on='key')
```

* 반대로 inner 옵션이 있다.
* 교집합처럼 공통된 요소만 가진다

```python
pd.merge(left, right, how='inner', on='key')
```

### 2. 파이썬의 대표 시각화 도구 -Matplotlib   56p

* 그래프를 그리는 모듈 : matplotlib.pyplot
* %matplotlib.inline : 그래프의 결과를 출력 세션에 나타나게 하는 설정

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure
plt.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,0])   # 총 18개
plt.show()
```

np.arrange 명령으로 0 부터 12까지 0.01. 간격으로 데이터를 만들고

그 리스트를 np.sin 에 입력하면 sin 값이 나옴 

```python
t = np.arange(0, 12, 0.01)
y = np.sin(t)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, y)
plt.grid()               # 격자 무늬
plt.xlabel('time')
plt.ylabel('Amplitude')  # Amplitude : 진폭
plt.title('Example of sinewave')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t))
plt.plot(t, np.cos(t))
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')  
plt.title('Example of sinewave')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')  
plt.title('Example of sinewave')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), lw=5, label='sin')          # lw 옵션으로 선의 굵기
plt.plot(t, np.cos(t), 'yellow', label='cos')         # 색상 지정 가능
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')  
plt.title('Example of sinewave')
plt.show()
```

* numpy 의 랜덤변수 함수를 이용해 데이터 세 개 만들기
* loc 옵션으로 평균값과 scale 옵션으로 표준편차를 지정

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)
```

```python
plt.figure(figsize=(10,6))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()   # label을 우측 상단에 보이게 함
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot((s1, s2, s3))
plt.grid()
plt.show()
```
