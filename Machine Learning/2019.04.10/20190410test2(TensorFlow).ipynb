```python
import tensorflow as tf
```

```python
xtrain = [1, 2, 3]
ytrain = [1, 2, 3]

# y = x * W + b
W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = xtrain * W + b
cost = tf.reduce_mean(tf.square(hypothesis - ytrain))    # reduce_mean   평균을 구함  # square 제곱을 구함
# optimizer
train = tf.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(cost)

train     # <tf.Operation 'GradientDescent' type=NoOp>
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for s in range(2000) :
    _, cost_val, W_val, b_val = sess.run([train, cost, W, b])
    #print(cost_val, W_val, b_val)
    if s % 20 == 0 :
        print(s, cost_val, W_val, b_val)
        
# 0 0.0015606364 [1.0457723] [-0.10405108]
```
