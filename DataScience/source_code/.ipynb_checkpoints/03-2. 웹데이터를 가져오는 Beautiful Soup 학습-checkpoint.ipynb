#### (연습) 네이터 영화 평점 기준 영화의 평점 변화 확인하기   158p

* [네이버 영화 2019년 7월 19일](https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20190719)

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from urllib.request import urlopen

url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20190719"

page = urlopen(url)   # 책에서는 이부분이 없는데 이걸 안 쓰면 안 나옴...

soup = BeautifulSoup(page, "html.parser")
print(soup.prettify())
```

```python
soup.find_all('div', 'tit5')
```

```python
# 영화 제목

soup.find_all('div', 'tit5')[0].a.string   # '알라딘'
```

```python
# 영화 평점

soup.find_all('td', 'point')[0].string   # '9.44'
```

```python
# 7월 1일부터 19일간으로 정의하고 그 날짜에 해당하는 영화 정보 전체를 찾아보기

date = pd.date_range('2019-7-1', periods=19, freq='D')

date
```

```python
import urllib
from tqdm import tqdm_notebook

movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date):
    html = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    
    response = urlopen(html.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))))
    
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})

movie.head()
```

```python
import numpy as np

movie_unique = pd.pivot_table(movie, index=["name"], aggfunc =np.sum)
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
```

* 왜 합계가 이렇게 나오지??? 짜증.... str 형이라 이렇게 나오나?

```python
tmp = movie.query('name == ["알라딘"]')
tmp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(loc='best')
plt.grid()
plt.show()
```

* 왜 또 y 축이 이렇게 나오는지.......

* plt.gca().invert_yaxis()   코드를 하면 y 축 reverse를 할 수 있다.

* 참고 : [frhyme님 github](https://frhyme.github.io/python-libs/plt_y_axis_reverse/)

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(loc='best')

# 추가 코드
plt.gca().invert_yaxis()

plt.grid()
plt.show()

```

* 위 코드를 실행하면 DataError: No numeric types to aggregate  오류가 발생한다.
