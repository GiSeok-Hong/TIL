```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv("Diabetes.csv")
data.head()
```

```python
sel_names = ['Pregnancies', 'Insulin', 'BMI', 'Age', 'Glucose', 'BloodPressure', 'DiabetesPedigreeFunction']
```

```python
x = data[sel_names]
y = data['Outcome']
type(x), type(y)    # (pandas.core.frame.DataFrame, pandas.core.series.Series)

x.shape, y.shape   # ((768, 7), (768,))

## 
x = x.values
y = y.values.reshape(len(y),1) # or y = y.values.reshape(-1,1)
# print(len(y))  # 768
x.shape, y.shape   # ((768, 7), (768, 1))
type(x), type(y)   # (numpy.ndarray, numpy.ndarray)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.3, random_state = 0)
xtrain
ytrain
```

```python
dstClf = DecisionTreeClassifier()
dstClf.fit(xtrain, ytrain)
```

```python
rResult = dstClf.predict(xtest)
rResult
```
