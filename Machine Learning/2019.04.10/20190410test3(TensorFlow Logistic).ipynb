```python
# Logistic Regression    로지스틱 회귀
# 분류에서 가장 단순한 모델로 2가지 중에 하나를 찾는 모델
# 참고 사이트  https://pythonkim.tistory.com/16?category=573319
#              https://pythonkim.tistory.com/22?category=573319
#              https://gdyoon.tistory.com/11
xdata = [
    [1,2],
    [2,5],
    [3,1],
    [4,6],
    [5,9],
    [6,8],
    [7,4]
]

ydata = [
    [0],
    [0],
    [1],
    [1],
    [0],
    [1],
    [0]
]

# x, y => 0 or 1
import tensorflow as tf
X = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W = tf.Variable(tf.random_normal([2,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
# Hypothesis sigmoid
# 시그모이드(sigmoid) 함수는 앞에서 배운 공식(Y = Wx + b )이 만들어 내는 값을
# 0과 1 사이의 값으로 변환한다.
# 어떤 값이든지 sigmoid  함수를 통과하기만 하면 0과 1 사이의 값이 되는 놀라운 기적을 보여준다.

# 가설
hypothesis = tf.sigmoid(tf.matmul(X,W) + b)   # matmul   matrix a 와 b를 곱해주는 함수?


# cost 함수 공식은 이러하니 외우거나 그렇구나~ 라고 넘어가자.   핸즈온 머신러닝책 190P 에 나와 있음
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))


# train 을 위한 식
train = tf.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(cost)
```

```python
# 예측
predicted = tf.cast(hypothesis > 0.5, dtype = tf.float32)
# h > 0.5 => 1,    h < 0.5 => 0

# 정확도
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for s in range(2000) :
    cost_val, _ = sess.run([cost, train], feed_dict={X:xdata, Y:ydata})
    if s % 200 == 0 :
        print(s, cost_val)
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X:xdata,Y:ydata})
    print("hypothesis 는 {0}, predicted 는 {1}, accuracy 는 {2}".format(h, c, a))
    print()
```
