```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
features = 2
niter = 2000
learning_rate = 0.01
```

```python
x = np.array([[0,0], [1,0], [0,1], [1,1]], np.float32)
```

```python
y = np.array([0, 1, 1, 0], np.float32).reshape(-1,1)
```

```python
#x.shape, y.shape    # ((4, 2), (4, 1))
```

```python
x1 = tf.placeholder('float', [None, 2])
y1 = tf.placeholder('float', [None, 1])
```

```python
#weight1 = tf.Variable(tf.random_normal([2,2]))
#weight2 = tf.Variable(tf.random_normal([2,1]))

weights = {
    "w_h1" : tf.Variable(tf.random_normal([2,2])),
    "w_out" : tf.Variable(tf.random_normal([2,1])) 
}
biases = {
    "b_h1" : tf.Variable(tf.zeros([2])),
    "b_out" : tf.Variable(tf.zeros([1]))
}
```

```python
def test_xor(x, weights, biases):
    hiddenLayer1 = tf.add(tf.matmul(x, weights['w_h1']), biases['b_h1'])
    hiddenLayer1 = tf.nn.sigmoid(hiddenLayer1)
    outLayer1 = tf.add(tf.matmul(x, weights['w_out']), biases['b_out'])
    return outLayer1
```

```python
model = test_xor(x, weights, biases)
```

```python
loss_func = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=model, labels=y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(loss_func)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for k in range(niter):
    tmp_cost, _ =sess.run([loss_func, optimizer], feed_dict={x1:x, y1:y})
    print(tmp_cost)
```
