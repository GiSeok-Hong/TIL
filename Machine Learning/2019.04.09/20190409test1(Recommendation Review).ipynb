```python
import os
import pandas as pd
from surprise import SVD   # 특이값 분해(Singular Value Decomposition, SVD)
from surprise import Dataset
```

```python
df = Dataset.load_builtin('ml-100k')
```

```python
data = pd.DataFrame(df.raw_ratings, columns = ['user', 'item', 'rate', 'id'])
```

```python
data.head()
```

```python
train_data = df.build_full_trainset()     # train set을 만들어줌
train_data   # <surprise.trainset.Trainset at 0x20690161240>
algo = SVD()
algo.train(train_data)       # <surprise.prediction_algorithms.matrix_factorization.SVD at 0x2068fcef940>
```

```python
test_data = train_data.build_testset()
predicts = algo.test(test_data)
from surprise import accuracy
```

```python
accuracy.rmse(predicts)
# RMSE: 0.6790        정확도가 0.67 정도라는 뜻
# 0.6789729336313879
```

```python
#######
from collections import defaultdict
def get_top_n(predictions, n = 10):
    top_n = defaultdict(list)
    for uid, iid, true_r, est,_ in predictions:
        top_n[uid].append((iid, est))
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key = lambda x: x[1], reverse = True)    # reverse = True 는 정렬
        top_n[uid] = user_ratings[:n]
    return top_n
```

```python
rResult = get_top_n(predicts)
```

```python
for uid, user_ratings in rResult.items():
    print(uid, [iid for (iid,_) in user_ratings])
```
